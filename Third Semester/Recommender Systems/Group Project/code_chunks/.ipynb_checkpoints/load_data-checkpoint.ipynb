```python
import os
import pandas as pd
```

```python
folder_path = './unzipped_csvs'
```

```python
# Empty DataFrame
table = pd.DataFrame()

for file_name in os.listdir(folder_path):
    file_path = os.path.join(folder_path, file_name)
    data = pd.read_csv(file_path, delimiter="\t", header=None)
    data=data.rename(columns={0: 'article_id', 1: 'Similarity'})
    
    # Extract title from file name
    title = os.path.splitext(file_name)[0]
    data["target_article"] = title
    
    # Append loaded data to the main table
    table = table.append(data)

table.set_index('target_article', inplace=True)
```

```python
print(table)
```

```python
table.to_csv("combined.csv")
```
