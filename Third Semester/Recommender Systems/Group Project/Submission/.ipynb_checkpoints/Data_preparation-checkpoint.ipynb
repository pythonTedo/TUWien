```python
newdf['authors'] = newdf['authors'].str.lower()
for i in range(0, len(newdf)):
    if ":" in newdf.authors[i]:
        newdf.authors[i] = newdf.authors[i].split(":")[-1].strip()[:-2]
newdf.authors = newdf.authors.str.replace("[", "").str.replace("]", "").str.replace("'", "").str.replace(" und", ",").str.replace(" &", ",")
newdf
```

## Retrieve the similarity values and enriching the article dataset

```python
# Empty DataFrame
ratings = pd.DataFrame()
folder_path = './unzipped_csvs'
for file_name in os.listdir(folder_path):
    file_path = os.path.join(folder_path, file_name)
    data = pd.read_csv(file_path, delimiter="\t", header=None)
    data=data.rename(columns={0: 'article_id', 1: 'Similarity'})
    
    # Extract title from file name
    title = os.path.splitext(file_name)[0]
    data["target_article"] = title
    
    # Append loaded data to the main table
    ratings = ratings.append(data)

#ratings.set_index('target_article', inplace=True)
```

```python
ratings = ratings.rename(columns={"target_article": "article_id", "article_id":"compared_article"})
```

```python
ratings.head()
ratings.shape
```

```python
ratings.to_csv("./data/rating_similarity.csv")
```

```python
merged = newdf.merge(right=ratings, how="left", on="article_id")
```
