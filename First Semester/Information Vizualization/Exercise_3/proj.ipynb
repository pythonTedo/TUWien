```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Exercise 3: Information vizualization

data source: https://www.kaggle.com/datasets/mrmorj/dataset-of-songs-in-spotify?select=genres_v2.csv

```python
data = pd.read_csv("genres_v2.csv")
data
```

```python
data.describe()
```

```python
plt.figure(figsize=(8,6))
data.genre.value_counts().plot.bar()
```

```python
sns.distplot(data.energy, kde="genre")
```

```python
ordered_days = data.genre.value_counts().index
g = sns.FacetGrid(data, col="key", row_order=ordered_days,
                  height=1.7, aspect=4,)
g.map(sns.kdeplot, "total_bill")
```
