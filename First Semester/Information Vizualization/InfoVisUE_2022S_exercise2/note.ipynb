```python
import pandas as pd
import json
from sklearn.decomposition import PCA
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
testData = pd.read_csv("static/data/betterLifeIndex_imputed.csv")
```

```python
pca = pd.read_csv("static/data/pca.csv")
```

```python
testData = testData.set_index("Countries")
```

```python
data_scaled = scaler.fit_transform(testData)
```

```python
data_scaled = pd.DataFrame(data_scaled, columns=testData.columns)
```

```python
pca = PCA(n_components=2)
data_pca = pd.DataFrame(pca.fit_transform(data_scaled), columns=['PCA Component One', 'PCA Component Two'])
```

```python
data_pca['country'] =  testData.index
```

```python
data_pca = data_pca.set_index('country')
```

```python
data_pca.to_csv("./static/data/pca.csv")
```
